```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
underlying = pd.read_csv("data/spx_minute_level_data_jan_mar_2024.csv")

end_date = datetime(2024, 3, 30)
```

```python
options : pd.DataFrame = pd.read_csv("data/cleaned_options_data.csv")
options["day"] = pd.to_datetime(options["ts_recv"].apply(lambda x: x.split("T")[0]))


options['datetime'] = pd.to_datetime(options['ts_recv'], format="%Y-%m-%dT%H:%M:%S.%fZ", utc=True).dt.tz_convert('US/Eastern')
options['symbol_data'] = options['symbol'].str.split('   ', expand=True)[1]

options['expiration_date'] = pd.to_datetime(options['symbol_data'].str[:6], format="%y%m%d")
options['type'] = options['symbol_data'].str[6]
options['strike'] = options['symbol_data'].str[7:].astype(float) / 1000

processed_options = options.drop(columns=['symbol_data'])
```

```python
processed_options.sort_values(by=['day', 'strike'])

current_date = datetime(2024, 1, 3)
processed_options = processed_options.loc[processed_options['expiration_date'] <= end_date, :] # only include options that expire before the end date

```
